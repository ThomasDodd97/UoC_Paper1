```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import rdDepictor
import rdkit
print(rdkit.__version__)

from PIL import Image
from io import BytesIO
def show_mol(d2d,mol,legend='',highlightAtoms=[]):
    d2d.DrawMolecule(mol,legend=legend, highlightAtoms=highlightAtoms)
    d2d.FinishDrawing()
    bio = BytesIO(d2d.GetDrawingText())
    return Image.open(bio)
# Buffer sets the distance apart the images are...
def show_images(imgs,buffer=0):
    height = 0
    width = 0
    for img in imgs:
        height = max(height,img.height)
        width += img.width
    width += buffer*(len(imgs)-1)
    res = Image.new("RGBA",(width,height))
    x = 0
    for img in imgs:
        res.paste(img,(x,0))
        x += img.width + buffer
    return res
```

# Poly(Glycerol Citrate) and Poly(Glycerol Citrate Itaconate)

```python
imgs = []

m = Chem.MolFromSmiles('*-OC(COC(=O)CC(CC(=O)O-*)C(=O)O-*)O-*')
d2d = Draw.MolDraw2DCairo(300,300)
dopts = d2d.drawOptions()
dopts.updateAtomPalette({8:(0.9215686274509803,0.49019607843137253,0.23529411764705882)})
imgs.append(show_mol(d2d,m,legend='Poly(Glycerol Citrate)'))

m = Chem.MolFromSmiles('*-OC(COC(=O)CC(CC(=O)O-*)C(=O)O-*)COC(=O)C(=C)CC(=O)O-*')
d2d = Draw.MolDraw2DCairo(300,300)
dopts = d2d.drawOptions()
dopts.updateAtomPalette({8:(0.9215686274509803,0.49019607843137253,0.23529411764705882)})
imgs.append(show_mol(d2d,m,legend='Poly(Glycerol Citrate Itaconate)'))

show_images(imgs)
```

# Reaction for Poly(Glycerol Citrate Itaconate)

```python
from rdkit.Chem import AllChem as AllChem
rxn = AllChem.ReactionFromSmarts('C(C(CO)O)O.C=C(CC(=O)O)C(=O)O.C(C(=O)O)C(CC(=O)O)(C(=O)O)O>>*-OC(COC(=O)CC(CC(=O)O-*)C(=O)O-*)COC(=O)C(=C)CC(=O)O-*.O',useSmiles=True)
d2d = Draw.MolDraw2DCairo(1600,600)
dopts = d2d.drawOptions()
dopts.useBWAtomPalette()
d2d.DrawReaction(rxn)
d2d.FinishDrawing()
bio = BytesIO(d2d.GetDrawingText())
Image.open(bio)
```
