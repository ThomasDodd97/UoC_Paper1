```python
X = df[["s1","s2","b1"]].to_numpy()
```

```python
y = df[["DeltaPolymerMass_pct"]].to_numpy()
```

```python
from pysr import PySRRegressor

model = PySRRegressor(
    maxsize=40,
    niterations=600,  # < Increase me for better results
    binary_operators=["+", "*"],
    unary_operators=[
        "exp",
        "inv(x) = 1/x",
        "sinh",
        "log",
        "cosh"
        # ^ Custom operator (julia syntax)
    ],
    extra_sympy_mappings={"inv": lambda x: 1 / x},
    # ^ Define operator for SymPy as well
    elementwise_loss="loss(prediction, target) = (prediction - target)^2",
    # ^ Custom loss function (julia syntax)
)

model1 = model.fit(X, y)
```

```python
fig = px.scatter_3d(df, x='s1', y='s2', z='b1', color='DeltaPolymerMass_pct')
fig.show()
```

```python
yPred_lis = model1.predict(X,index=30)
df["yPred_lis"] = yPred_lis
fig = px.scatter_3d(df, x='s1', y='s2', z='b1', color='yPred_lis')
fig.show()
```
