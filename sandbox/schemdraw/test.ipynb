# ALTERNATIVE TUTORIAL

```python
import schemdraw
from schemdraw import flow

with schemdraw.Drawing() as d:
    d.config(fontsize=11)
    b = flow.Start(w=1,h=1).label('Start')
    flow.Arrow().down(d.unit/4).label(r'$i=1$', loc='bottom')
    d1 = flow.Decision(w=4,h=2.4,E='Y',S='N').label('Sample Limit, ' + r'$p_{tot}$' + ',\n Reached?\n(' + r'$i=p_{tot}$' + '?)', ofst=[0.0,-0.2])
    flow.Arrow().down(d.unit/4)
    d2 = flow.Decision(w=3.5,h=2.4,E='Y',S='N').label(f'Prior-Forming\nDataset, {r'$\mathbb{D}$'},\nExtant?', ofst=[0,-0.1])
    flow.Arrow().down(d.unit/4)
    p1 = flow.Box(w=2.5, h=1).label('Quasirandom\nSampling')
    flow.Arrow().down(d.unit/4).label(r'$[\mathbf{x}]$', loc='top').label(r'$i=i+p_{init}$', loc='bottom')
    io1 = flow.Data(w=3.2, h=1).label('Output: Sample(s)\nto be Evaluated')
    flow.Arrow().down(d.unit/4).label(r'$[n_{Ci},n_{It}]$', loc='bottom')
    p7 = flow.Box(w=4, h=1).label('Sample(s) Manufactured\nand Tested')
    flow.Arrow().down(d.unit/4).label(r'$[\Delta m_{\%}]$', loc='bottom')
    p6 = flow.Box(w=3.7, h=1.2).label('New Predictor(s) and\nTarget(s) Merged\ninto Dataset, ' + r'$\mathbb{D}$')
    flow.Arrow().left(d.unit/4).at(p6.W)
    flow.Wire('|-', arrow='->').to(d1.W).label(r'$\mathbb{D}=[\mathbf{X},\mathbf{\Delta m_{\%}}]$', loc='left', rotate=90, ofst=-0.5)

    flow.Arrow().right(d.unit/4).at(d1.E)
    p2 = flow.Start(w=1, h=1).anchor('W').label('End')

    flow.Arrow().right(d.unit/2.5).at(d2.E)
    # p3 = flow.Box(w=1.7, h=1.4).label('Gaussian\nProcess\nPrior\nFormed')
    flow.Arrow().down(d.unit/1.55)
    p4 = flow.Box(w=1.7, h=1).label('Bayesian\nSampling')

    flow.Wire('|-', arrow='->').at(p4.S).to(io1.E).label(r'$[\mathbf{x}]$', loc='top', ofst=[0.0,-1.3]).label(r'$i=i+p_{iter}$', loc='bottom', ofst=[0.0,0.0])
    d.save('OptimisationAlgorithm.png',transparent=False,dpi=1000)
```
