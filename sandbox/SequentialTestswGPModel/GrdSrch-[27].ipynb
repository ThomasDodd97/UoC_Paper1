```python
xgb_model = xgb.XGBRegressor()
xgb_model.load_model('/Users/thomasdodd/Library/CloudStorage/OneDrive-MillfieldEnterprisesLimited/Cambridge/PhD/writing/papers/UoC_Paper1/Sandbox/Modelling/ModelMk2.json')
y_pred = xgb_model.predict(X)
df2 = pd.DataFrame({'s1': X[:, 0],'s2': X[:, 1],'b1': X[:, 2], 'y_pred': y_pred})
df2["y_pred"] = y_pred
fig = px.scatter_3d(df2, x='s1', y='s2', z='b1', color='y_pred')
fig.show()
```
