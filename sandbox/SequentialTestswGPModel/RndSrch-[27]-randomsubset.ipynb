```python
trials = 1000

X_lis = []
for i in range(trials):
    NumberOfPriorSamplesToDraw = 8
    SampleIndexes_lis = random.sample(df.index.values.tolist(),27)
    X_sublis = []
    for j in SampleIndexes_lis:
        X_sublis.append([df.iat[j,0],df.iat[j,1],df.iat[j,2]])
    X_lis.append(X_sublis)

y_max_lis = []
for X in X_lis:
    y_pred_lis = []
    for row in X:
        s1 = row[0]
        s2 = row[1]
        b1 = row[2]
        y_pred_lis.append(client.predict([{"s1":s1,"s2":s2,"b1":b1}])[0]["t1"][0])
    y_max_lis.append(np.max(np.array(y_pred_lis)))

y_max_arr = np.array(y_max_lis)
print(y_max_arr.tolist())
print(np.average(y_max_arr))
```
